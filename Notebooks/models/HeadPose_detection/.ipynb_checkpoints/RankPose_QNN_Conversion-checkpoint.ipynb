## Processing output

```python
np.set_printoptions(precision=2)

target_size = 224
batch_size = 16
n_class = 3
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
dataset = laod_dataset(data_type='AFLW2000', base_dir="./assets", filename="./assets/input.txt", target_size=target_size, n_class=n_class, split='val')
test_loader = DataLoader(dataset, batch_size=batch_size, shuffle=False, num_workers=1, pin_memory=True)
```

```python
def draw_3d_axis(img, yaw, pitch,roll, size=100):
    yaw = math.radians(yaw)
    pitch = math.radians(pitch)
    roll = math.radians(roll)
    
    length = size//2
    R_yaw = np.array([[math.cos(yaw), -math.sin(yaw),0],
                      [math.sin(yaw),math.cos(yaw),0],
                      [0,0,1]])
    
    R_pitch = np.array([[math.cos(pitch),0, math.sin(pitch)],
                       [0,1,0],
                      [-math.sin(pitch),0,math.cos(pitch)]])
                      
    R_roll = np.array([[1,0,0],
                     [0,math.cos(roll), -math.sin(roll)],
                       
                      [0,math.sin(roll),math.cos(roll)]])
    
    x_axis = np.dot(np.dot(R_roll,R_pitch), np.dot(R_yaw, np.array([length,0,0])))
    y_axis = np.dot(np.dot(R_roll,R_pitch), np.dot(R_yaw, np.array([0,length,0])))
    z_axis = np.dot(np.dot(R_roll,R_pitch), np.dot(R_yaw, np.array([0,0,length])))
    
    center = (img.shape[1] //2,img.shape[0] //2)
    
    x_end = (int(center[0] + x_axis[0]), int(center[1] - x_axis[1]))
    y_end = (int(center[0] + y_axis[0]), int(center[1] - y_axis[1]))
    z_end = (int(center[0] + z_axis[0]), int(center[1] - z_axis[1]))
    
    cv2.line(img,center,x_end,(0,0,255),2)
    cv2.line(img,center,y_end,(0,255,0),2)
    cv2.line(img,center,z_end,(255,0,0),2)
    return img
```

```python
def put_text_on_image(img,yaw,pitch,roll):
       
    cv2.putText(img, f"Yaw:{yaw: .2f}",(20,40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0),2)
    cv2.putText(img, f"Pitch:{pitch: .2f}",(20,80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0),2)
    cv2.putText(img, f"Roll:{roll:.2f}",(20,120), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0),2)

    #draw_direction(img,ypr_pre[0],ypr_pre[1])
    draw_3d_axis(img,yaw,pitch,roll,size=100)
    return
    
    
    
```

```python
import math
maxsize=0
os.environ['base_dir']="./assets/"
base_dir ="./assets/"
dst_pred = "./assets/output_pre/"
dst_lbl = "./assets/output_lbl/"


if os.path.exists(dst_pred):
    exit(True)
else:
    os.mkdir("./assets/output_pre/")
    
if os.path.exists(dst_lbl):
    exit(True)
else:
    os.mkdir("./assets/output_lbl/")

if os.path.exists("./assets/output_dsp_pre"):
    exit(True)
else:
    os.mkdir("./assets/output_dsp_pre/")


if os.path.exists("./assets/output_cpu_pre"):
    exit(True)
else:
    os.mkdir("./assets/output_cpu_pre/")

res = 0
folders =["./assets/output_dsp/","./assets/output_cpu/"]
for i, (images, labels) in enumerate(test_loader):
        images = images.to(device)
        labels_np = labels.detach().numpy()
        labels = labels.to(device)
        
        preds = model(images)
        preds_np = preds.detach().cpu().numpy()

        diff_total = calculate_diff(preds, labels, mean=False)
        if res ==30:
            break
    
        for j in range(len(images)):
            diff = diff_total[j]
            #diff_list.append(diff)
            
            if n_class == 4:
                ypr_lbl = quat2euler(*labels_np[j])
                ypr_pre = quat2euler(*preds_np[j])
            else:
                ypr_lbl = labels_np[j]
                ypr_pre = preds_np[j]
            
            #print("ypr lbl===== ", ypr_lbl)
            #print(" ypr_pred ==========",ypr_pre)
                
            img_path = os.path.join(base_dir,dataset.img_paths[i*batch_size+j])
            img= cv2.imread(img_path)
            
            
            
            put_text_on_image(img,ypr_pre[0],ypr_pre[1],ypr_pre[2])
            draw_3d_axis(img,ypr_pre[0],ypr_pre[1],ypr_pre[2],size=100)
            output_path = os.path.join(dst_pred,f"image_{i*batch_size+j}.jpg")
            cv2.imwrite(output_path,img)
            
            img_path = os.path.join(base_dir,dataset.img_paths[i*batch_size+j])
            img= cv2.imread(img_path)

            put_text_on_image(img,ypr_lbl[0],ypr_lbl[1],ypr_lbl[2])
            draw_3d_axis(img,ypr_lbl[0],ypr_lbl[1],ypr_lbl[2],size=100)
            output_path = os.path.join(dst_lbl,f"image_{i*batch_size+j}.jpg")
            cv2.imwrite(output_path,img)

      
            
            
            for folder in folders:
                if res==30:
                    break
                
                folder = os.path.join(folder, f"Result_{res}","_501.raw")
                with open(folder, 'rb') as f:
                    raw_output = np.fromfile(f,dtype=np.float32)
                    #print(raw_output.shape)
                
                ypr = raw_output.reshape((1,3))
                ypr =ypr.astype(np.float32)
      
                img_path = os.path.join(base_dir,dataset.img_paths[i*batch_size+j])
                img= cv2.imread(img_path)
                put_text_on_image(img,ypr[0][0],ypr[0][1],ypr[0][2])
                
                res_img = draw_3d_axis(img,ypr[0][0],ypr[0][1],ypr[0][2],size=100)
                #print(type(res_img))
                if "dsp" in folder:
                    #print(folder)
                    output_path = os.path.join("./assets/output_dsp_pre",f"image_{res}.jpg")
                    cv2.imwrite(output_path,res_img)

                else:
                    output_path = os.path.join("./assets/output_cpu_pre",f"image_{res}.jpg")
                    cv2.imwrite(output_path,res_img)
            res =res +1     
                
            '''
            
            
            #print("[ %d/%d ]" % (i*batch_size + j + 1, total_size), end='\r')'''
            if maxsize > 0 and (i*batch_size + j + 1) >= maxsize:
                break
print("\n[ INFO ] TEST end")
```
